## Imports

```python
import pandas as pd
```

## Creating new data with row 1 as header (skip the metadata row 0)

```python
df = pd.read_excel('../data/raw/default_of_credit_card_clients.xls', header=1, index_col=0)
df.head()
print('-'*100)
print(df.info())
```

## Checking for Missing Values and Duplicates

```python
print("Missing Values:")
print(df.isnull().sum())
print("-"*100)
print("\nDuplicates:")
print(df.duplicated().sum())
```

## Dropping Duplicates

```python
df_cleaned = df.drop_duplicates()
print(f"Duplicates after dropping: {df_cleaned.duplicated().sum()}")
```

## Exporting as Processed Data ready for Analysis

```python
df_cleaned.to_csv('../data/processed/cleaned_default_data.csv', index=False)
```
